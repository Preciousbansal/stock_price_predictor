splitting data into training and testing 

```python
data_training = pd.DataFrame(df['Close'][0:int(len(df)*0.70)])  #70% data training
data_testing = pd.DataFrame(df['Close'][int(len(df)*0.70):int(len(df))])  #30% data is testing

print(data_training.shape)
print(data_testing.shape)
```

```python
data_training.head()
# first 70% of data
```

```python
data_testing.head()
# rest 30% data
```

scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array = scaler.fit_transform(data_training)
data_training_array
```

```python
data_training_array.shape
```

```python
x_train=[]    #we will take first 'x' days values and on the basis of that we'll predict the coming value
y_train=[]    # stores the value to be predicted (predicted class)
# hum jo bhi value predict karna chahte hai, vo hum y_train se lenge aur fir jab uss y_train value ka prediction ho jayega toh utni values x_train se hatt jayengi aur y_train ki saari values x_train m store ho jayengi aur fir vo x_train ka rpart hongi bas.

for i in range(100,data_training_array.shape[0]):
  x_train.append(data_training_array[i-100:i])
  y_train.append(data_training_array[i,0])

x_train,y_train = np.array(x_train),np.array(y_train)
#converted x_train and y_train into numpy arrays
```

```python
x_train.shape
```

training model

```python
from keras.layers import Dense, Dropout, LSTM
from keras.models import Sequential
```

```python
model = Sequential()

#1st layer
model.add(LSTM(units =50, activation = 'relu', return_sequences=True, input_shape=(x_train.shape[1],1)))
model.add(Dropout(0.2))

#2nd layer
model.add(LSTM(units =50, activation = 'relu', return_sequences=True))
model.add(Dropout(0.3))

#3rd layer
model.add(LSTM(units =80, activation = 'relu', return_sequences=True))
model.add(Dropout(0.4))

#4th layer
model.add(LSTM(units =120, activation = 'relu'))
model.add(Dropout(0.5))


model.add(Dense(units=1))   # units =1 because we only want to predict one value
#dense layer connects all above LSTM layers
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train, y_train, epochs=50)
```

```python
model.save('keras_model.h5')
```
